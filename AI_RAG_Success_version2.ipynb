```python
from sentence_transformers import SentenceTransformer

# Initialize the SentenceTransformer model for embedding
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Path to the JSON file (your file that contains multiple objects)
json_file_path = "/data.json"

# Open and read the JSON file
with open('/content/AI_RAG_Data_3.json', 'r') as f:
    data = json.load(f)

# Dictionary to hold embeddings for each entry in the JSON file
embeddings = {}

# Iterate through the key-value pairs in the JSON data
for key, value in data.items():
    # Assuming each value is a dictionary or object to be embedded
    if isinstance(value, dict) or isinstance(value, list):
        # Flatten the object to a string representation for embedding
        flattened_text = json.dumps(value)
    else:
        # If value is a simple string or number, convert it to a string
        flattened_text = str(value)

    # Generate the embedding for the flattened text
    embedding = model.encode([flattened_text])[0].tolist()  # Convert NumPy array to list

    # Store the embedding in the dictionary with the original key
    embeddings[key] = {
        "embedding": embedding,
        "original_data": value
    }

# Write the embeddings to a new JSON file in a neat and tidy format
with open('embeddings_output.json', 'w+') as f:
    json.dump(embeddings, f, indent=4, separators=(',', ': '))
```
