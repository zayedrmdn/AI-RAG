```python
import json
import os
from sentence_transformers import SentenceTransformer
import numpy as np

# Initialize the SentenceTransformer model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Function to get embedding for a text
def get_embedding(text):
    # Convert the input text into a list (if it's a single string)
    if isinstance(text, str):
        text = [text.replace('\n', ' ')]

    # Generate embeddings using SentenceTransformer
    embeddings = model.encode(text)

    # Return the first embedding as a Python list (convert from NumPy array)
    return embeddings[0].tolist()

sentence = "How do I configure GPIO pins for I2C communication on STM32, and what code is needed for setting up the clock speed to 100kHz? format the response in markdown format"
get_embedding(sentence)
```

```python
# Path to the JSON file
json_file_path = '/content/data.json'

# Open and read the JSON file
with open(json_file_path, 'r') as f:
    data = json.load(f)

# Dictionary to hold embeddings for each entry in the JSON file
embeddings = {}

# Assuming the JSON structure is a dictionary or list, loop through the entries
for key, value in data.items():
    # If the value is a string, generate an embedding for it
    if isinstance(value, str):
        embeddings[key] = get_embedding(value)
    elif isinstance(value, dict) or isinstance(value, list):
        # If the value is more structured, flatten it for embedding
        flattened_text = json.dumps(value)
        embeddings[key] = get_embedding(flattened_text)

# Write the embeddings to a new JSON file
with open('embeddings.json', 'w+') as f:
    json.dump(embeddings, f)
```

```python
import json
import numpy as np
from sentence_transformers import SentenceTransformer
from gradio_client import Client

# Load SentenceTransformer model for embedding
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Load embeddings from a JSON file
embeddings = json.load(open('embeddings.json', 'r'))

# Initialize Gradio Client for Hugging Face model (replacing ChatTemplate)
client = Client("yuntian-deng/ChatGPT")


# Function to calculate cosine distance between two vectors
def cosine_distance(a, b):
    return 1 - sum([a_i * b_i for a_i, b_i in zip(a, b)]) / (
        sum([a_i ** 2 for a_i in a]) ** 0.5 * sum([b_i ** 2 for b_i in b]) ** 0.5
    )


# Function to generate embeddings for input text using SentenceTransformers
def get_embedding(text):
    return embedding_model.encode([text])[0]  # Single sentence embedding


# Find the nearest embedding to the given query embedding
def nearest_embedding(embedding):
    nearest, nearest_distance = None, 1

    for path, embedding2 in embeddings.items():
        distance = cosine_distance(embedding, embedding2)
        if distance < nearest_distance:
            nearest, nearest_distance = path, distance

    return nearest


# Main loop to interact with the user and generate responses
while True:
    prompt = input('user: ')
    if prompt == "exit":
        break

    # Get the embedding of the user's input
    query_embedding = get_embedding(prompt)

    # Find the nearest embedding in the dataset
    context = nearest_embedding(query_embedding)

    # Read the context from the nearest match
    with open(context, 'r') as f:
        data = f.read()

    # Use the Hugging Face model to generate a response
    response = client.predict(
        inputs=f"Here are some facts that can help you answer the following question: {data}\n{prompt}",
        api_name="/predict"
    )

    # Print the response from the model
    print(f"assistant: {response}")
```
