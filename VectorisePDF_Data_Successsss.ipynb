```python
import os
from PyPDF2 import PdfReader
from Crypto.Cipher import AES


# Function to extract text from a single PDF file
def extract_text_with_pyPDF(PDF_File):
    pdf_reader = PdfReader(PDF_File)
    raw_text = ''
    for i, page in enumerate(pdf_reader.pages):
        text = page.extract_text()
        if text:
            raw_text += text
    return raw_text

# Function to process all PDF files in a directory
def process_pdfs_in_directory(pdf_directory, output_directory):
    # Ensure the output directory exists
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Loop through all files in the PDF directory
    for pdf_file in os.listdir(pdf_directory):
        if pdf_file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_directory, pdf_file)

            # Extract the text from the PDF
            extracted_text = extract_text_with_pyPDF(pdf_path)

            # Define the output text file path
            output_file_path = os.path.join(output_directory, pdf_file.replace('.pdf', '.txt'))

            # Save the extracted text to the output file
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(extracted_text)

            print(f"Extracted and saved text from {pdf_file} to {output_file_path}")

# Specify the input directory with PDFs and the output directory for extracted text files
pdf_directory = "/content/AI-RAG-PDF_directory/AI-RAG-PDF"
output_directory = "/content/extracted_texts"

# Process all PDFs in the directory
process_pdfs_in_directory(pdf_directory, output_directory)
```

```python
import fitz  # PyMuPDF
import os
from PIL import Image

# STEP 1: Specify the directory containing PDF files and the directory to save the images
pdf_directory = "/content/AI-RAG-PDF_directory/AI-RAG-PDF"
image_save_directory = "/content/Extracted_Images"

# Ensure the save directory exists
os.makedirs(image_save_directory, exist_ok=True)

# STEP 2: Iterate through all PDF files in the directory
for pdf_file_name in os.listdir(pdf_directory):
    if pdf_file_name.endswith(".pdf"):  # Process only PDF files
        pdf_file_path = os.path.join(pdf_directory, pdf_file_name)
        pdf_file = fitz.open(pdf_file_path)

        # STEP 3: Iterate over PDF pages and extract images
        for page_index in range(len(pdf_file)):

            # Get the page itself
            page = pdf_file.load_page(page_index)
            image_list = page.get_images(full=True)

            # Print the number of images found on this page
            if image_list:
                print(f"[+] Found a total of {len(image_list)} images on page {page_index} in {pdf_file_name}")
            else:
                print(f"[!] No images found on page {page_index} in {pdf_file_name}")

            for image_index, img in enumerate(image_list, start=1):
                # Get the XREF of the image
                xref = img[0]

                # Extract the image bytes
                base_image = pdf_file.extract_image(xref)
                image_bytes = base_image["image"]

                # Get the image extension (e.g., png, jpeg)
                image_ext = base_image["ext"]

                # Create a unique name for the image based on the PDF file, page, and image index
                image_name = f"{os.path.splitext(pdf_file_name)[0]}_page_{page_index+1}_img_{image_index}.{image_ext}"

                # Define the full path to save the image
                image_path = os.path.join(image_save_directory, image_name)

                # Save the image to the specified directory
                with open(image_path, "wb") as image_file:
                    image_file.write(image_bytes)
                    print(f"[+] Image saved as {image_path}")

print("[+] All images extracted and saved.")
```
